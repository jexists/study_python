# Pandas
파이썬에서 사용하는 데이터 분석 라이브러리

```python
import pandas as pd
```

# 1. Series
- 1차원 데이터 (정수, 실수, 문자열 등)
- 라벨이 붙은 1차원 배열 형태
- 단일 데이터 타입의 데이터를 다룰 때 사용

## Series 객체 생성

예) 1월부터 4월까지 평균 온도 데이터 (-20, -10, 10, 20)

```python
temp = pd.Series([-20, -10, 10, 20])
print(temp)
```

```python
temp[0] # 1월 온도
```

```python
temp[2] # 3월 온도
```

# Series 객체 생성 (Index 지정)

```python
temp = pd.Series([-25, -15, 15, 25], index=['Jan', 'Feb', 'Mar', 'Apr'])
print(temp)
```

```python
temp['Jan'] # index Jan(1월)에 해당하는 데이터 출력
```

```python
temp['Apr'] # index Apr(4월)에 해당하는 데이터 출력
```

```python
temp['Apri'] # 존재하지 않는 Index 접근 시도 에러
```

```python
temp[0]
```

```python
temp[10] # 존재하지 않는 Index 접근 시도 에러
```
