# 스타일

```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'AppleGothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y, linewidth=15)
```

## 마커 (marker)

https://matplotlib.org/stable/api/markers_api.html

```python
plt.plot(x,y, marker='o')
```

```python
plt.plot(x,y, marker='o', linestyle='')
```

```python
plt.plot(x,y, marker='o', linestyle='None')
```

```python
plt.plot(x,y, marker='v')
```

```python
plt.plot(x,y, marker='v', markersize=10)
```

```python
plt.plot(x,y, marker='x', markersize=10)
```

```python
plt.plot(x,y, marker='X', markersize=10)
```

```python
plt.plot(x,y, marker='X', markersize=10, markeredgecolor='red')
```

```python
plt.plot(x,y, marker='o', markersize=10, markeredgecolor='red')
```

```python
plt.plot(x,y, marker='o', markersize=20, markeredgecolor='red')
```

```python
plt.plot(x,y, marker='o', markersize=20, markeredgecolor='red', markerfacecolor='yellow')
```

## 축약어

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

```python
# mfc: markerfacecolor, ms: markersize, mec: markeredgecolor, mew: markeredgewidth ls: linestyle
plt.plot(x,y, marker='o', mfc='red', ms=10, mec='blue', mew=5, ls=':')
```

## 투명도



```python
# alpha 투명도 (0~1)
plt.plot(x,y, marker='o', mfc='red', ms=10, alpha=0.3)
```
