```python
N = df.shape[0]
N
```

```python
index = np.arange(N)
index
```

```python
# w = 0.25
# plt.bar(index - ?)

w = 0.25
plt.bar(index - w, df['국어'])
```

```python
# w = 0.25
# plt.bar(index - ?)

w = 0.25
plt.bar(index - w, df['국어'])
plt.bar(index, df['영어'])
```

```python
# w = 0.25
# plt.bar(index - ?)

w = 0.25
plt.bar(index - w, df['국어'])
plt.bar(index, df['영어']) 
plt.bar(index + w, df['수학'])
```

```python
# w = 0.25
# plt.bar(index - ?)

w = 0.25
plt.bar(index - w, df['국어'], width=w)
plt.bar(index, df['영어'], width=w) 
plt.bar(index + w, df['수학'], width=w)
```

```python
# w = 0.25
# plt.bar(index - ?)

w = 0.25
plt.bar(index - w, df['국어'], width=w, label='국어')
plt.bar(index, df['영어'], width=w, label='영어') 
plt.bar(index + w, df['수학'], width=w, label='수학')
plt.legend()
```

```python
# w = 0.25
# plt.bar(index - ?)

w = 0.25
plt.bar(index - w, df['국어'], width=w, label='국어')
plt.bar(index, df['영어'], width=w, label='영어') 
plt.bar(index + w, df['수학'], width=w, label='수학')
plt.legend(ncol=3)
```

```python
# w = 0.25
# plt.bar(index - ?)

w = 0.25
plt.bar(index - w, df['국어'], width=w, label='국어')
plt.bar(index, df['영어'], width=w, label='영어') 
plt.bar(index + w, df['수학'], width=w, label='수학')
plt.legend(ncol=3)
plt.xticks(index, df['이름'])
```

```python
plt.figure(figsize=(10, 5))
plt.title('학생별 성적')
w = 0.25
plt.bar(index - w, df['국어'], width=w, label='국어')
plt.bar(index, df['영어'], width=w, label='영어') 
plt.bar(index + w, df['수학'], width=w, label='수학')
plt.legend(ncol=3)
plt.xticks(index, df['이름'], rotation=60)
plt.show()
```
